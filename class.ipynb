```python
import pandas as pd
```

## Series
A Series is a one-dimensional array-like object containing a series of values (similar to a NumPy array). It may also have an array of values called the index.

```python
# Creating a Pandas Series
series = pd.Series([3, 2, -8, 5])
series_with_index = pd.Series([3, 2, -8, 5], index=['a', 'b', 'c', 'd'])

print("Series with default index:\n", series)
print("Series with custom index: \n", series_with_index)
```

## DataFrame
A DataFrame is a data structure used to represent multidimensional data, mostly tabular data. If we have a table represented as a DataFrame, we can consider each column of the table as a Series.

A common way to construct a DataFrame is using a dictionary:

```python
my_dict = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
}
my_df = pd.DataFrame(my_dict)
my_df
```

## Reading CSV files using Pandas

```python
df = pd.read_csv("https://raw.githubusercontent.com/amith-c/pandas-intro/main/sample.csv")
```

 We created a Pandas DataFrame `df` that now contains the data read from the file `sample.csv`.

 The file has the following columns:
 1. rollno
 2. name
 3. mark1
 4. mark2
 5. mark3

 If we try printing the value of `df`:

```python
print(df)
```

## How do we access values?
### Accessing entire columns
If we wanted to access the entire `name` column, we would do something like this:

```python
df['name']
```

To access more than one column from the DataFrame, we can pass a list of column names into the square bracket like so:

```python
df[['rollno', 'name']]
```

### Accessing a single row
We can use the `iloc` or `loc` properties of the DataFrame to access specific rows.

```python
# Accessing the 2nd row of 'df'
print(df.iloc[1])

# Accessing rows 3 to 6 (index 2 to 6)
print(df.iloc[2:6])
```

## Commonly asked functions

### Sorting values
We sort values in the DataFrame using the `sort_values()` function. This function takes a sort key as parameter and returns the sorted Dataframe. 

```python
print("Names before sorting: ")
print(df['name'])

# Sort the DataFrame
df = df.sort_values("name", ascending=True)

print("\nNames after sorting: ")
print(df['name'])
```

### Displaying records
We use two functions to view records from the beginning and end of the DataFrame, which are `head()` and `tail()` respectively. `head()` returns the first 5 records in the DataFrame by default. But if we need more than 5, we can simply pass the number to the function as a parameter.

```python
# Print the first 5 records
print(df.head()) 

# Print the first 7 records
print(df.head(7))

# Print the last 5 records
print(df.tail()) 

# Print the last 3 records
print(df.tail(3))
```

### Special Functions
Some special functions that might be useful for the exams:

```python
# We first retrieve the 'mark1' column in the DataFrame
mark1 = df['mark1']

# Average (Mean)
print("Average of 'mark1': ", mark1.mean())

# Median
print("Median of 'mark1': ", mark1.median())

# Standard Deviation
print("Standard deviation of 'mark1': ", mark1.std())

# Maximum value
print("Highest value of 'mark1': ", mark1.max())

# Minimum value
print("Lowest value of 'mark1': ", mark1.min())
```

### Filtering rows
Suppose we need to obtain the details of the students having more than 20 marks in 'mark1'. We filter the DataFrame using the condition `df['mark1'] > 20` inside the square brackets instead of a column name.

```python
filtered_df = df[df['mark1'] > 20]
print("Filtered DF: ", filtered_df)
```

And if we want to find the average of the marks in this filtered dataframe, we can simply call the `mean()` function on it.

```python
filtered_df['mark1'].mean()
```

## Writing to CSV files
Finally, we get to the part where we learn to write data into CSV files. Consider the following data:

```python
fields = ["Name", "Branch", "Year", "CGPA"]
rows = [["Nikhil", 'CSE', 2, 8.0],
        ['Sanchit', 'CSE', 2, 9.1],
        ['Aditya', 'IT', 2, 9.3]]
```

We have two lists - `fields` containing the headers and `rows` containing the data.

To create a new DataFrame using this data, we use the `pd.DataFrame()` function:

```python
new_df = pd.DataFrame(rows, columns=fields)

# Print the DF
print(new_df)
```

Here, the `columns` parameter indicates the names of the columns in the DataFrame. If we omit this parameter, the columns would simply be named 0, 1, 2, etc.

### Saving the CSV file
Now let's see how we can write this DataFrame to a CSV file. We can simply call the `to_csv()` function:

```python
new_df.to_csv("file.csv")
```
